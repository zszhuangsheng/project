## Data Collection and Wragling

winemag-data_first150k.csv contains 10 columns and 150k rows of wine reviews.

- A country: The country that the wine is from
- B description: A few sentences from a sommelier describing the wine's taste, smell, look, feel, etc.
- C designation: The vineyard within the winery where the grapes that made the wine are from
- D points: The number of points WineEnthusiast rated the wine on a scale of 1-100 (though they say they only post reviews for wines that score >=80)
- E price: The cost for a bottle of the wine
- F province: The province or state that the wine is from
- G region_1: The wine growing area in a province or state (ie Napa)
- H region_2: Sometimes there are more specific regions specified within a wine growing area (ie Rutherford inside the Napa Valley), but this value can sometimes be blank
- I variety: The type of grapes used to make the wine (ie Pinot Noir)
- J winery: The winery that made the wine

```python
import csv

filename="winemag-data_first150k.csv"
with open(filename) as f:
    reader=csv.reader(f)
    header_row=next(reader)
    
    description=[]
    for row in reader:
        description.append(row[2])  ##extract the data of country
    text=" ".join(description)
```

```python
# import libraries
import matplotlib.pyplot as plt
import os
from os import path
from PIL import Image
from matplotlib import pyplot as plt
from scipy.misc import imread
import random
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import numpy as np
```

```python
def wc_english():
    # get the the current path of file 
    d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
    # get the background picture
    background_Image = np.array(Image.open(path.join(d, "mask1900.jpg")))
    # or
    # background_Image = imread(path.join(d, "mask1900.jpg"))
    # extract the color of background picture 
    img_colors = ImageColorGenerator(background_Image)
    # set the stop words
    stopwords = set(STOPWORDS)
    stopwords.add('now')
    wc = WordCloud(
        margin = 10, # set the margin
        mask = background_Image,
        width=1500,
        height=960,
        scale = 2,
        max_words = 200, # set the maximum number of words
        min_font_size = 4, 
        stopwords = stopwords,
        random_state = 42,
        background_color = 'white', # set the background colour
        max_font_size = 200, # set the size of maximum word 
        )
    wc.generate_from_text(text)
    # wc.generate(text)
    # set background color according to picture
    wc.recolor(color_func=img_colors)
    #save 
    wc.to_file('1900pro1.png')
    # show the wordcould
    plt.imshow(wc,interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
wc_english()
```

### world distribution

```python
import csv
import pyecharts
from pyecharts import Map

filename="winemag-data_first150k.csv"
with open(filename) as f:
    reader=csv.reader(f)
    header_row=next(reader)
    
    countries=[]
    for row in reader:
        countries.append(row[1])  ##只提取country的数据

##去除列表中重复的元素             
countrycopy=list(set(countries))
```

```python
##统计列表中国家重复的次数，并以字典的形式呈现
country_dict={}
for country in countries:
     if country in countrycopy:
            country_dict[country] = country_dict.get(country,0)+1

            country_dict["United States"]=62397
country_dict.pop("US")
country_dict.pop("")
country_dict
##再将作出的字典变成一个元素为字典的列表
data = []
for key, value in country_dict.items():
    data.append({'name': key, 'value': value})

districts = list(country_dict.keys())  ##将字典的key变成一个命名为districts的列表
areas = list(country_dict.values())    ##与上一步同样的道理
world_map = Map('The distribution of countries', width=1200, height=600)
world_map.add("The distribution of countries", districts, areas, maptype='world', is_visualmap=True, visual_range=[min(areas), max(areas)],
               visual_text_color='#000', is_map_symbol_show=False, is_label_show=False)
world_map.render(path="worldmap.html")#生成html文件
```

### The distribution of US

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py

df=pd.read_csv("winemag-data_first150k.csv",index_col=0)  # 0 means select the index cow
states=df[df['country']=='US']['province'].value_counts().reset_index()
states_ref = pd.read_csv('states.csv')
states_ref.head()
```

```python
states_all = states.merge(states_ref, how='inner', left_on='index', right_on='States')
for col in states_all.columns:
    states_all[col] = states_all[col].astype(str)
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
states_all['text'] = states_all['index'] + '<br>' +\
    'Numer '+states_all['province']
data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = states_all['Abbreviation'],
        z = states_all['province'].astype(float),
        locationmode = 'USA-states',
        text = states_all['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Millions USD")
        ) ]
layout = dict(
        title = ' US Agriculture Exports by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

In United States, California has many wine factories.

```python
import folium
```

```python
CA1=df[df['province']=='California']["region_1"].value_counts().reset_index()
# we try to find top 10 county i 
CA1_top_10=CA1[0:12]
CA1_top_10.drop(CA1_top_10.index[[2]],inplace=True)
CA1_top_10["new_coord"]=[[35.1979,-118.2224],[38.4833, -122.8166],[35.6405,-120.6800],[38.29, -122.46],[38.4,-123.0],[38.2555,-122.3513],[38.6999,-122.9666],[34.4208,-119.6981],[38.75, -122.9],[34.6349,-120.3154],[38.3666, -122.4999]]
from colour import Color
def get_color_range(n, output_type='hex'):
    red = Color('red')
    blue = Color('blue')
    color_range = list(red.range_to(blue, n))
    if output_type == 'hex':
        return [c.get_hex_l() for c in color_range]
    else:
        return [c.get_rgb() for c in color_range]
county_codes = list(CA1_top_10["index"])
color_range = get_color_range(len(county_codes))
code2color = {county_code:c for county_code, c in zip(county_codes, color_range)}
CA_map = folium.Map(location=[36.778259, -119.417931], zoom_start=5)
for _, row in CA1_top_10.iterrows():
    
    site_name = row["index"]
    number_of_winery = row["index"]
    lat, long = row.new_coord
    
    folium.RegularPolygonMarker([lat, long],
        popup='{} / numer of winery: {}'.format(site_name, row.region_1),
        fill_color=code2color[number_of_winery],
        number_of_sides=20,
        radius=8
        ).add_to(CA_map)

CA_map
```
